```python
import pandas as pd
df=pd.read_csv("/content/spam.csv",encoding="ISO-8859-1,")
df.head()
```

```python
df.drop(["Unnamed: 2"],axis=1,inplace=True, errors='ignore')
df.drop(["Unnamed: 3"],axis=1,inplace=True, errors='ignore')
df.drop(["Unnamed: 4"],axis=1,inplace=True, errors='ignore')
```

```python
df['v1']=df['v1'].map({'ham':1,'spam':-1})
df.head()
```

```python
Spam=df.v2
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer

sw=SnowballStemmer("english")
tk=TweetTokenizer()
```

```python
Spam=Spam.apply(lambda x:" ".join([sw.stem(i.lower()) for i in tk.tokenize(x)]))
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words("english")

sw
```

```python
Spam=Spam.apply(lambda x:" ".join([i for i in x.split() if i not in sw]))
Spam
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x=tfidf.fit_transform(Spam)
print(x)
print()
print(tfidf.get_feature_names_out())
```

```python
df['v1']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,df['v1'],test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_predict=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_predict)*100)
```
